```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
np.random.seed(42) # Set seed for reproducibility
```

```python
# Simulation for 1d
# Parameters
n = 100
a, b, c = 2.0, 3.0, 1.5
```

```python
# Create correlated X and Z
X = np.random.normal(0, 1, n)
Z = 0.5 * X + np.random.normal(0, 1, n)
e = np.random.normal(0, 1, n)
```

```python
# Full model: Y = a + bX + cZ + e
Y_full = a + b * X + c * Z + e
```

```python
# Model with both predictors
X_full = sm.add_constant(np.column_stack((X, Z)))
model_full = sm.OLS(Y_full, X_full).fit()
```

```python
# Model with X only
X_partial = sm.add_constant(X)
model_partial = sm.OLS(Y_full, X_partial).fit()
```
