```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# Simulation for 2b
np.random.seed(42)
```

```python
# Generate clean data
n = 100
X = np.random.normal(0, 1, n)
e = np.random.normal(0, 1, n)
Y_clean = 2 + 3 * X + e
```

```python
# Add outliers to Y
Y_outlier = Y_clean.copy()
Y_outlier[[10, 20, 30]] += 15  # Add large values to simulate outliers
```

```python
# Fit models
X_sm = sm.add_constant(X)
model_clean = sm.OLS(Y_clean, X_sm).fit()
model_outlier = sm.OLS(Y_outlier, X_sm).fit()
```

```python
# Return coefficients from both models
model_clean.params, model_outlier.params
```
