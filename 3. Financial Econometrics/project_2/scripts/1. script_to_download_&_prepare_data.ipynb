```python
import pandas as pd
import yfinance as yf
from fredapi import Fred

fred = Fred(api_key='********************')

# 1. Download US macro (CPI, PPI, Core CPI)
series = {
    'CPI': 'CPIAUCSL',
    'PPI': 'PPIACO',
    'CoreCPI': 'CPILFESL'
}
for name, sid in series.items():
    df = fred.get_series(sid, observation_start='2005-01-01', observation_end='2024-12-31')
    df = df.to_frame(name)
    df.index = pd.to_datetime(df.index)
    df.to_csv(f'{name.lower()}.csv')
    print(f"Saved {name.lower()}.csv")

# 2. Download Brent & WTI from Yahoo Finance
for ticker, name in [('BZ=F','Brent'), ('CL=F','WTI')]:
    df = yf.download(ticker, start='2000-01-01', end='2024-12-31', interval='1mo')
    df = df[['Close']].rename(columns={'Close': name})
    df.to_csv(f'{name.lower()}-monthly.csv')
    print(f"Saved {name.lower()}-monthly.csv")

# 3. Combine Brent & WTI
br = pd.read_csv('brent-monthly.csv', index_col=0, parse_dates=True)
wt = pd.read_csv('wti-monthly.csv',   index_col=0, parse_dates=True)
df_bt = pd.concat([br, wt], axis=1).dropna()
df_bt.to_csv('brent_wti.csv')
print("Saved combined brent_wti.csv")
```
